## Using Kmeans to Quantize Colors

*Quantizing colors means we'll reduce the number of unique colors here to K unique colors. Let's try just 3 colors!*

```python
(h,w,c) = image_as_array.shape 
```

```python
h
```

```python
w
```

```python
c
```

### Convert from 3d to 2d

Kmeans is designed to train on 2D data (data rows and feature columns), so we can reshape the above strip by using (h,w,c) ---> (h * w,c)

```python
image_as_array2D = image_as_array.reshape(h*w,c)
```

```python
image_as_array2D.shape
```

```python
from sklearn.cluster import KMeans 
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
labels = kmeans.fit_predict(image_as_array2D)
```

```python
labels
```

## Obtaining new rgb codes of the colors that we'll see

```python
rgb_codes = kmeans.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_img=np.reshape(rgb_codes[labels],(h,w,c))
```

# Quantized_img

```python
plt.figure(figsize=(12,6))
plt.imshow(quantized_img)
```

## CONCLUSION :

Since the number of colors we want is equal to the 'k' value that we choose therefore we can play around with different
values of k and get different forms of images.
